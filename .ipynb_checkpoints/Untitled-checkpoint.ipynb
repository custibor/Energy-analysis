```python
import datetime
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df = pd.read_csv("data/PJME_hourly.csv")
df.head()
```

```python
def get_year(x):
    date_info = x.strip().split(" ")[0]
    return  int(date_info.split("-")[0])
def get_month(x):
    date_info = x.strip().split(" ")[0]
    return int(date_info.split("-")[1])
def get_hour(x):
    date_info = x.strip().split(" ")[1]
    return int(date_info.split(":")[0])

def get_dayOfWeek(x):
    dt = datetime.datetime.strptime(x.strip().split(" ")[0],"%Y-%m-%d")
    return int(dt.weekday())
```

```python
df["Year"] = df["Datetime"].apply(lambda x: get_year(x))
df["Month"] = df["Datetime"].apply(lambda x: get_month(x))
df["Hour"] = df["Datetime"].apply(lambda x: get_hour(x))
df["DayOfWeek"] = df["Datetime"].apply(lambda x: get_dayOfWeek(x))
```

```python
df.head()
```

```python
yearly_data = df[["PJME_MW","Year"]].groupby("Year").agg(["mean","std","median",lambda x: np.percentile(x,0.25),lambda x: np.percentile(x,0.75)])
```

```python
yearly_data.columns = yearly_data.columns.droplevel()
yearly_data = yearly_data.reset_index()
```
